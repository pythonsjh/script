```python
#3
# -*- coding: utf-8 -*-
class Counter:
    def __init__(self,count,step=1):#요구사항1 step의 default는 1
        self.count=count#count, step초기화
        self.step=step    
    def __str__(self):#요구사항2
        return  "[Count (step : %d)]%d"%(self.step,self.count)
    def incr(self):#요구사항3
        self.count=self.count+self.step
    def __call__(self):#요구사항4
        self.incr()
    def __add__(self, other): #요구사항5   
        return Counter(self.count+other,1) #주의. 인스턴스형식으로 리턴해야줘야된다       
    def __sub__(self, other): #요구사항5           
        return Counter(self.count-other,2) #주의. 인스턴스형식으로 리턴해야줘야된다
    def __cmp__(self, other):#요구사항6
        return self.count - other
    
c = Counter(10) 
d = Counter(10, 2) 

c.incr()
d.incr()
print c
print d

c()
d()
print c
print d

c=c+5
d=d-5
print c
print d

print c > 10 
print d > 10 
print c < 10 
print d < 10 
print c == 17    # 이퀄연산자는 반환값이 0이어야 true가된다
print d != 9    # notequal연산자는 반환값이 0이면 false가 된다
```
